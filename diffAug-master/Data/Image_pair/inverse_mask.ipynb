```python
from PIL import Image
import os
import numpy as np

mask_dir = './masks'
output_dir = './InvertedMasks'
os.makedirs(output_dir, exist_ok=True)

def is_binary(image_array):
    """Check if the image array is binary (only contains 0 and 255 values)."""
    unique_values = np.unique(image_array)
    return set(unique_values).issubset({0, 255})

def invert_and_check_binary(mask_path):
    """Invert the mask and check if it's binary."""
    mask = Image.open(mask_path).convert('L')
    mask_np = np.array(mask)
    inverted_mask_np = 255 - mask_np
    inverted_mask = Image.fromarray(inverted_mask_np)

    # if still binary
    binary_check = is_binary(inverted_mask_np)

    return inverted_mask, binary_check

for filename in os.listdir(mask_dir):
    if filename.endswith(('.png', '.jpg', '.jpeg')):
        mask_path = os.path.join(mask_dir, filename)

        inverted_mask, is_binary_mask = invert_and_check_binary(mask_path)

        output_path = os.path.join(output_dir, filename)
        inverted_mask.save(output_path)

        print(f"{filename}: Binary - {is_binary_mask}")
```
